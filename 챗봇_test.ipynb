```python
## 문제
## 문서번호 "HR-2025-030" 에서
## * 부서: "HR"
## * 연도: "2025"
## * 번호: "030"
## 를 슬라이싱만으로 추출하라.
```

```python
문서번호 = "HR-2025-030"
부서 = 문서번호[0:2]
연도 = 문서번호[3:7]    
번호 = 문서번호[8:11] 
print(f"부서: {부서}")  # 출력: HR
print(f"연도: {연도}")  # 출력: 2025   
print(f"번호: {번호}")  # 출력: 030
```

https://www.coupang.com/np/search?component=&q=%EB%9D%BC%EB%A9%B4&channel=user

"%EB%9D%BC%EB%A9%B4&" => **유니코드**

```python
url = "https://www.coupang.com/np/search?component=&q=%EB%9D%BC%EB%A9%B4&channel=user"
```

```python
import urllib.parse 
decode = urllib.parse.unquote(url)
print(f"디코딩된 URL: {decode}")  # 출력: https://www.coupang.com/np/search?component=&q=라면&channel=user
```

```python
import urllib.parse

# 'decode' 변수에서 '라면' 부분을 추출
start_index = decode.find('q=') + 2
end_index = decode.find('&channel=')
korean_text = decode[start_index:end_index]

# 한글 텍스트를 유니코드로 변환
encode = urllib.parse.quote(korean_text)
print(f"인코딩된 URL 부분: {encode}")  # 출력: %EB%9D%BC%EB%A9%B4
```
