```python
from openpyxl import *
```

```python
# 엑셀파일 생성(만들기)
wb = Workbook() # 워크북객체 생성
wb.save('테스트엑셀파일.xlsx')
```

```python
# 기존 엑셀파일 로드 (원래 있던거)
wb = load_workbook('테스트.xlsx')
```

```python
# 시트 확인
wb.sheetnames
```

```python
# 시트 로드
ws = wb.active  # 활성화된 시트 로드
# ws = wb['questions']  # 시트 이름으로 로드
# ws = wb.worksheets[0]  # 인덱스로 로드
```

```python
ws['A1'] = 'Hello'
ws.cell(2,2, '안녕??')
ws.append(['1', '2', '3', '4', '5'])
```

```python
# 저장
wb.save('테스트엑셀파일.xlsx')
```

```python
# 데이터 확인
ws['A1'].value
```

```python
# 데이터 확인 => 데이터 없음
ws['B1'].value
ws.cell(2,2).value
```

```python
for i in range(2):
    print(ws.cell(i+1, 1).value)
```

```python
# 행 갯수 확인
ws.max_row
```
