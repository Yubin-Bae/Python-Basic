# 챗봇 사용 연습
# .JSON 파일 사용

```python
# Sales.json 파일을 json으로 불러오기
import json
with open("Sales.json", "r", encoding="utf-8-sig") as file:
    sales_data = json.load(file)
    print(sales_data)
```

```python
sales_data # 리스트 형태([ 로 시작)
```

```python
# 총 판매 금액 누적
total = 0                

# 가장 많이 팔린 상품 정보 저장
best_name = ""                            
best_qty = 0

# 상품 하나씩 꺼내기
for s in sales_data:

    # 판매 금액 계산해서 누적
    total += s["price"] * s["qty"]

    # 지금 상품이 더 많이 팔렸는지 확인
    if s["qty"] > best_qty:

        # 최고 기록이면 교체
        best_qty = s["qty"]
        best_name = s["name"]

print("총 판매 금액:", f"{total:,}원")
print("베스트셀러:", best_name)


# 이 코드가 가장 쉬운 이유
# 변수 수: 최소
# 반복문: 1개
# 함수: 사용X
# 리스트/딕셔너리: 기본 접근만 사용
```
